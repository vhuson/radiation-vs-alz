# Evaluate merged dataset through plotting

Let's make a nice scatterplot where country is coded

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
dataset_path = "datasets/"
df = pd.read_csv(dataset_path + "merged_alz_sol.csv")

df.head()
```

```python


# Scatter plot with seaborn, using 'color' for coloring
plt.figure(figsize=[10,10])
sns.scatterplot(x='Avg Theoretical Potential', y='YearsLifeLost_T', hue='Country Solar', data=df)

# Adding labels and title
plt.xlabel('Solar Power')
plt.ylabel('Years of Life lost from Alzheimers (per 100,000)')
#plt.title('Scatter Plot with Color')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
# Show the plot
plt.show()
```

```python
# Scatter plot but for GDP this time
plt.figure(figsize=[10,10])
sns.scatterplot(x='GDP (USD 2018)', y='YearsLifeLost_T', hue='Country Solar', data=df)

# Adding labels and title
plt.xlabel('GDP')
plt.ylabel('Years of Life lost from Alzheimers (per 100,000)')
#plt.title('Scatter Plot with Color')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
# Show the plot
plt.show()
```

```python
# Scatter plot but GDP vs solar power
plt.figure(figsize=[10,10])
sns.scatterplot(x='GDP (USD 2018)', y="Avg Theoretical Potential", hue='Country Solar', data=df)

# Adding labels and title
plt.xlabel('GDP')
plt.ylabel('Solar Power')
#plt.title('Scatter Plot with Color')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
# Show the plot
plt.show()
```
